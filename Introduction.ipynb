```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('C', 'G')])
```

```python
G
```

```python
plt.axis('off')
nx.draw_networkx(G, node_size=600, font_color='white')
plt.show()
```

```python
DG = nx.DiGraph()
```

```python
DG.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('C', 'G')])
```

```python
DG
```

```python
plt.axis('off')
nx.draw_networkx(DG, node_size=600, font_color='white')
plt.show()
```

```python
WG = nx.Graph()
WG.add_edges_from([('A', 'B', {"weight": 10}),
                    ('A', 'C', {"weight": 20}),
                   ('B', 'D', {"weight": 30}),
                   ('B', 'E', {"weight": 40}),
                   ('C', 'F', {"weight": 50})])
labels = nx.get_edge_attributes(WG, 'weight')
```

```python
plt.axis('off')
nx.draw_networkx(WG, pos=nx.spring_layout(WG, seed=0), node_size=600, font_color='white', font_size=14, cmap='coolwarm')
nx.draw_networkx_edge_labels(WG, pos=nx.spring_layout(WG, seed=0), edge_labels=labels)
plt.show()
```

```python
labels
```

```python
G1 = nx.Graph()
G1.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('C', 'G')])
plt.axis('off')
nx.draw_networkx(G1, node_size=600, font_color='white')
plt.show()

print(f"Is graph 1 connected? {nx.is_connected(G1)}")
```

```python
G2 = nx.Graph()
G2.add_edges_from([('A', 'B'), ('A', 'C'), ('D', 'E'), ('D', 'F')])
plt.axis('off')
nx.draw_networkx(G2, node_size=600, font_color='white')
plt.show()
print(f"Is graph 2 connected? {nx.is_connected(G2)}")
```
